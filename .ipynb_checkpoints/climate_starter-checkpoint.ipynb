```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func , inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base=automap_base()
# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement=Base.classes.measurement
Station=Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session=Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the columns in the two tables so we can find what queries to run starting with measurement
inspector = inspect(engine)
measurement_columns = inspector.get_columns("measurement")
for c in measurement_columns:
    print(c['name'],c['type'])
```

```python
#Find the columns in the station table
station_columns = inspector.get_columns("station")
for c in station_columns:
    print(c['name'],c['type'])
```

```python
# Find the most recent date in the data set.
most_recent_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()[0]
print(most_recent_date)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
last_date = dt.datetime.strptime(most_recent_date, '%Y-%m-%d')
# Calculate the date one year from the last date in data set.
start_date = dt.date(last_date.year -1, last_date.month, last_date.day)
precipitation = session.query(Measurement.date,Measurement.prcp).filter(Measurement.date >= start_date).\
    order_by(Measurement.date).all()

# Perform a query to retrieve the data and precipitation scores
#for p in precipitation:
    #print(p)

# Save the query results as a Pandas DataFrame. Explicitly set the column names
precipitation_df = pd.DataFrame(precipitation)
precipitation_df.columns = ["Date","Precipitation"]
clean_precipitation=precipitation_df.dropna()

# Sort the dataframe by date
clean_precipitation=clean_precipitation.set_index("Date")


# Use Pandas Plotting with Matplotlib to plot the data
clean_precipitation.plot()
plt.ylim(0,7)
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Inches')
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
clean_precipitation.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
number_stations = session.query(Station.station).count()
number_stations
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
active_stations = session.query(Measurement.station,func.count(Measurement.station)).\
    order_by(func.count(Measurement.station).desc()).\
    group_by(Measurement.station).all()

print(active_stations)
    
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
#Using the essentially the same code as above the most active station is the first in descending order Station USC00519281
most_active_station = session.query(Measurement.station,func.count(Measurement.station)).\
    order_by(func.count(Measurement.station).desc()).\
    group_by(Measurement.station).first()

most_active_station_number = most_active_station[0]




#lowest temperature
lowest_temp = session.query(func.min(Measurement.tobs)).\
    filter(Measurement.station == most_active_station_number).all()
print(lowest_temp)

#Highest temperature
highest_temp = session.query(func.max(Measurement.tobs)).\
    filter(Measurement.station == most_active_station_number).all()
print(highest_temp)

#Average temperature
average_temp = session.query(func.avg(Measurement.tobs)).\
    filter(Measurement.station == most_active_station_number).all()
print(average_temp)
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

most_recent_station_date = session.query(Measurement.date,Measurement.tobs).\
    filter(Measurement.station == most_active_station_number).\
    order_by(Measurement.date.desc()).first()[0]
#most_recent_station_date

#most recent reading for the most active station is 2017-08-18 
mst_rec_stn_dt = dt.datetime.strptime(most_recent_station_date, '%Y-%m-%d')
stn_start_date = dt.date(mst_rec_stn_dt.year-1, mst_rec_stn_dt.month, mst_rec_stn_dt.day)
twelve_month_data = session.query(Measurement.date,Measurement.tobs).\
    filter(Measurement.station == most_active_station_number).\
    filter(Measurement.date>= stn_start_date).all()

twelve_month_tobs_data = pd.DataFrame(twelve_month_data)['tobs']
#twelve_month_tobs_data

hist = twelve_month_tobs_data.hist(bins=12)
plt.xlabel = "Temperature"
plt.ylabel = "frequency"
plt.ylim(0,70)
```
